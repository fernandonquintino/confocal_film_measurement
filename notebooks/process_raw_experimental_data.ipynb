```python
df = pd.concat(dfs, axis=0).reset_index(drop=True)
df.reset_index(drop=True, inplace=True)
```

```python
df['date'] = ''
for i in range(len(list_dates)):
    df.loc[df['df']==(i+1), 'date'] = list_dates[i]

df['number'] = df.groupby('date').cumcount() + 1

df['date_identifier'] = (
    df['date']
    + '_'
    + df['number'].astype(str).str.zfill(2)
)
```

```python
new_pattern_dict = {'churn-annular': 'churn',
'annular': 'annular',
'bubbly-churn': 'dispersed',
'wavy-annular': 'wavy',
'slug': 'slug',
'churn': 'churn',
'plug': 'plug',
'wavy': 'wavy',
'stratified': 'stratified',
'bubbly': 'bubbly',
'dispersed bubbles': 'dispersed',
'bubble': 'plug',
'stratified-wavy': 'wavy',
'annular-wavy': 'wavy'}

df['reduced_pattern'] = df['pattern'].map(new_pattern_dict)
```

```python
df.loc[df['Ql']<5, 'sensor'] = 'q1'
```

```python
for i in range(df.shape[0]):
    if df.loc[i, 'sensor'] in ['q1', 'q1_v2', 'q1_v3']:
        df.loc[i, 'Ql'] = 0.93*df.loc[i, 'Ql']
    if df.loc[i, 'sensor'] in ['q2']:
        df.loc[i, 'Ql'] = 0.95*df.loc[i, 'Ql']
```

```python
df['Ptp'] = 3.5*df['dp_exp'] + Po
df['Ttp'] = (df['Tl'] + df['Tg'])/2
```

```python
df['jl'] = df['Ql']/(A*60000)
df['jg'] = df['Qg']/(A*60000)
df['dpdz_tp'] = df['dp_exp']/L
```

```python
df = df.drop(columns=['Pnovo', 'df', 'date', 'number'])
df = df[
    [
        'Tl', 'Tg', 'Ttp',
        'Ql', 'Qg', 'jl', 'jg',
        'Pg', 'Ptp',
        'dp_exp', 'dpdz_tp',
        'date_identifier', 'pattern', 'reduced_pattern'
    ]
]
df = df.round({
    'Tl': 1, 'Tg': 1, 'Ttp': 1,
    'Ql': 2, 'Qg': 2, 'jl': 2, 'jg': 2,
    'Pg': 0, 'Ptp': 0, 'dp_exp': 0, 'dpdz_tp': 0,
})

df['Pg'] = df['Pg'].astype(int)
df['Ptp'] = df['Ptp'].astype(int)
df['dp_exp'] = df['dp_exp'].astype(int)
df['dpdz_tp'] = df['dpdz_tp'].astype(int)
```

```python
df = df.reset_index().rename(columns={'index': 'run_id'})
df['run_id'] = df['run_id'] + 1
```

```python
save_path = Path.cwd().parent / 'data/silver_layer'
save_path.mkdir(parents=True, exist_ok=True)

df.to_pickle(
    save_path / 'experimental_results.pkl',
)

df.to_csv(
    save_path / 'experimental_results.csv'
)
```
