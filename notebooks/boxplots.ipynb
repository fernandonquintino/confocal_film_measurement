# Experimental conditions

```python
silver_path = Path.cwd().parent / 'data/silver_layer/'
gold_profile_path = Path.cwd().parent / 'data/gold_layer/boxplots'
```

```python
df_confocal = pd.read_pickle(silver_path / 'confocal_results.pkl')
df_confocal_processed = pd.read_pickle(silver_path / 'confocal_results_processed.pkl')
df_experimental = pd.read_pickle(silver_path / 'experimental_results.pkl')
```

```python
CM = 1 / 2.54

FIG_WIDTH_CM = 8.4
FIG_HEIGHT_CM = 6.8

FIG_SIZE = (
    FIG_WIDTH_CM * CM,
    FIG_HEIGHT_CM * CM
)
```

```python
reduced_pattern = ['bubbly', 'dispersed', 'plug', 'slug', 'wavy']
all_patterns = ['annular', 'bubbly', 'churn', 'dispersed', 'plug', 'slug', 'stratified', 'wavy']
```

# Unprocessed

```python
missing_threshold = 100
df = pd.merge(df_experimental, df_confocal, on='run_id', how='inner')
df_restricted = df.loc[df['t2_missing_percentage']<missing_threshold]
```

```python
fig1, ax = boxplot_plot(df_restricted, 't1_missing_percentage', 'Missing percentage', '\%', patterns=reduced_pattern)
fig2, ax = boxplot_plot(df_restricted, 't2_missing_percentage', 'Missing percentage', '\%', patterns=reduced_pattern)
fig3, ax = boxplot_plot(df_restricted, 't1_mean', 'Mean thickness', 'mm', patterns=reduced_pattern)
fig4, ax = boxplot_plot(df_restricted, 't2_mean', 'Mean thickness', 'mm', patterns=reduced_pattern)
```

```python
fig1f, ax = boxplot_plot(df, 't1_missing_percentage', 'Missing percentage', '\%', patterns=all_patterns)
fig2f, ax = boxplot_plot(df, 't2_missing_percentage', 'Missing percentage', '\%', patterns=all_patterns)
fig3f, ax = boxplot_plot(df, 't1_mean', 'Mean thickness', 'mm', patterns=all_patterns)
fig4f, ax = boxplot_plot(df, 't2_mean', 'Mean thickness', 'mm', patterns=all_patterns)
```

```python
fig_path1 = gold_profile_path / f'unprocessed_percentage1.eps'
fig_path2 = gold_profile_path / f'unprocessed_percentage2.eps'
fig_path3 = gold_profile_path / f'unprocessed_thickness1.eps'
fig_path4 = gold_profile_path / f'unprocessed_thickness2.eps'

fig1f.savefig(fig_path1, format='eps', bbox_inches='tight')
fig2f.savefig(fig_path2, format='eps', bbox_inches='tight')
fig3f.savefig(fig_path3, format='eps', bbox_inches='tight')
fig4f.savefig(fig_path4, format='eps', bbox_inches='tight')
```
